```python
def read_target(file):
    targets = pd.read_excel(file)
    targets = targets[targets['x'] != 'x'].astype(float)
    return targets
```

```python
def read_rawimage(file):
    df = pd.read_excel(file,header=None)
    image_cols = df.iloc[:,:-4].astype(str)
    last_col = df.iloc[:,-2]
    image = image_cols.apply(lambda col: np.array(eval(','.join(col.to_list()))).astype(float),axis=1)
    new_df = pd.DataFrame({'Image':image,'Unknown':last_col})
    return new_df
```

```python
img = read_rawimage('data-RawImage-rpi1.xlsx')
```

```python
def view_top_slices(df,n_slices=5,figsize=(12,3),method=np.median):
    fig,ax = plt.subplots(1,n_slices,figsize=figsize,sharex=True,sharey=True)
    topvals = df.groupby('Unknown')['Unknown'].count().sort_values(ascending=False) ## find most common values in c3
    for i,v in enumerate(topvals.index[:n_slices]):
        filt = df[df['Unknown'].round() == np.round(v)] ## grab all images for c3 value
        imgs = np.array([im for im in filt['Image']]) ## create array of images at c3 value
        stack = method(imgs,axis=0) ## median stack all images
        ax[i].imshow(stack)
        ax[i].set_title(np.round(v,2))
    plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
view_top_slices(img)
```

```python
import matplotlib.animation as animation
from IPython.display import HTML
def make_animation(df,interval=50,savefig=False):
    fig = plt.figure()
    ims = []
    for i in df['Image']:
        im = plt.imshow(i,animated=True)
        ims.append([im])
    ani = animation.ArtistAnimation(fig,ims,interval=interval,blit=True,repeat_delay=1000)
    return ani
```

```python
ani = make_animation(img)
```

```python
HTML(ani.to_html5_video())
```

```python
ani.save('time_series_clip.mp4')
```
