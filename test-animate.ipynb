```python
from IPython.display import HTML
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

```python
import pandas as pd
```

```python
def read_data(file):
    f = open(file)
    lines = f.readlines()
    f.close()
    df = pd.DataFrame([eval(line) for line in lines],columns=['image','c1','c2','c3','c4'])
    return df
```

```python
clip = read_data('data-RawImage-fullMag-rpi1.csv')
noclip = read_data('data-RawImage-emptyMag-rpi1.csv')
```

```python
def make_animation(df,interval=50,savefig=False):
    fig = plt.figure()
    ims = []
    for i in df.image:
        im = plt.imshow(np.array(i),animated=True)
        ims.append([im])
    ani = animation.ArtistAnimation(fig,ims,interval=interval,blit=True,repeat_delay=1000)
    if not savefig:
        HTML(ani.to_html5_video())
    else:
        ani.save(savefig)
```

```python
make_animation(clip,savefig='clip.mp4')
make_animation(noclip,savefig='noclip.mp4')
```
